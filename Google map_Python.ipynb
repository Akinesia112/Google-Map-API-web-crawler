## google maps 爬蟲 & 距離計算 & 顯示熱門時段 function

```python
#找尋範圍內地點
def gmap_nearby_search(location, radius, Type):
    place_list = []
    results = maps.places_nearby(location=location, radius=radius,type=Type)
    
    #處理回傳資料(地點名、經緯、ID、評價、評價數)
    while 1:
        json_data = results
        for place in json_data['results']:
            
            p_name = place['name']
            p_lat = place['geometry']['location']['lat']
            p_lng = place['geometry']['location']['lng']
            p_id = place['place_id']
            if 'rating' in place:
                p_rating = place['rating']
                p_total_ratings = place['user_ratings_total']
            else:
                p_rating = -1
                p_total_ratings = -1
  
            place_list.append([p_name, p_lat, p_lng, p_rating, p_total_ratings, p_id])
    
        print('Already save %d locations.'%(len(place_list)))
        
        if 'next_page_token' in json_data:
            time.sleep(3) #google api delay
            results = maps.places_nearby(page_token=json_data['next_page_token'])
        else:
            break
            
    return place_list

#計算經過所有地點之路徑長及交通時間
def get_total_distance(location_lists):
    maps = googlemaps.Client(key=KEY)
    results = maps.distance_matrix(location_lists, location_lists)
    #加總路徑及時間
    total_distance = 0
    total_duration = 0
    for i in range(len(location_lists)-1):
        total_distance += results['rows'][i]['elements'][i+1]['distance']['value']
        total_duration += results['rows'][i]['elements'][i+1]['duration']['value']
    
    return total_distance, total_duration ,results

#時段人流
def get_popular_times(place_id):
    try:
        result = populartimes.get_id(KEY, place_id)
        popular_times = sum([np.array(t['data']) for t in result['populartimes']])
        fig = sns.barplot(x=list(range(0,24)), y=popular_times, color='salmon').set_title("Popular hour times")
        plt.xlabel("Hours")
        plt.ylabel("Week Popularity")
        plt.show()
        return popular_times
    except:
        print('This place_id no popular times information.')
```
